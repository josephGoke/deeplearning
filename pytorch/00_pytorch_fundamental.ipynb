### Reshaping, Viewing, Stacking, squeezing & unsqueezing Tensors

* Reshaping - reshapes an input tensor to a defined shape
* View - Return a view of an input tensor of certain shape but keep the same memory as the original tensor
* Stacking - combine multiple tensors on top of each other (Vstack)
* Squeeze - remove all '1' dimensions from a tensor
* Unsqueeze - add a '1' dimension to a target tensor
* Permute - Return a view of the input with dimensions permuted(swapped) in a certain way

```python
X = torch.arange(1.,10.)
print(X, X.shape)
```

```python
x_reshaped = X.reshape(1, 9)
x_reshaped, x_reshaped.shape
```

```python
# Change the view
z = X.view(1, 9)    # does the same thing as X.reshape, except that it shares the same memory with the original tensor
z, z.shape
```

```python
# Changing z changes x
z[:, 0] = -2
z, X
```

```python
# Stacking tensors
x_stacked = torch.stack([X, X, X, X], dim=1)
x_stacked
```

```python
#torch.hstack stacks tensors in sequence horizontally(column-wise)
x_stacked = torch.hstack([X,X,X,X])
x_stacked
```

```python
# torch.vstack stacks tensors in sequence vertically(row-wise)
x_stacked = torch.vstack([X,X,X,X])
x_stacked
```

```python
# torch.squeeze() - removes all single dimensions from a target tensor
print(f'x_reshaped: {x_reshaped}')
print(f'shape: {x_reshaped.shape}')

x_squeezed = x_reshaped.squeeze()
print(f'x_squeezed: {x_squeezed}')
print(f'Shape: {x_squeezed.shape}')
```

```python
    # torch.unsqueeze()
x_unsqueezed = x_squeezed.unsqueeze(dim=0)
x_unsqueezed
```

```python
#   torch.permute() - rearranges the dimensions of a target to tensor in a specified order
x = torch.randn(2,3,5)
print(x)
print(x.size())

x_permute = torch.permute(x, dims=(2,0,1))
print(f'Permuted tensor: {x_permute}')
print(f'Size: {x_permute.size()}')
```
