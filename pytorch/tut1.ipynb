## Tensors: Multidimensional array(fundamental data structure in PyTorch)

```python
tensor_1d = torch.tensor([1, 2, 3])
print("1D Tensor (Vector):")
print(tensor_1d)
print()
a = torch.ones(3)
print('1-D Tensor: ', a)
# print(a[2])
print()

tensor_2d = torch.tensor([[1, 2], [3, 4]])
print("2D Tensor (Matrix):")
print(tensor_2d)
print()

random_tensor = torch.rand(2,3, dtype=float)
print('Random Tensor (2x3):')
print(random_tensor)
```

```python
zeros_tensor = torch.zeros(2, 3)
print("Zeros Tensor (2x3):")
print(zeros_tensor)
print()

ones_tensor = torch.ones(2, 3)
print("Ones Tensor (2x3):")
print(ones_tensor)

## Creating tensor using shape dimension
shape = (2,3,)
tensor = torch.rand(shape)
print(tensor)
print(f'Tesnor shape: {tensor.shape}')
print(f'Tesnor datatype: {tensor.dtype}')
print(f'Tesnor device: {tensor.device}')
```

#### Tensor Operations(Indexing, Slicing, Reshaping)

```python
##  Indexing &Slicing
print(random_tensor)
elements = random_tensor[0,:]
print(f'Sliced Tensor (first row): {elements}')

print('')

print(random_tensor)
element = random_tensor[0,1]
print(f'indexed element (first row, second column): {element}')
```

```python
## Reshaping
print(random_tensor)
print(random_tensor.size())
print('')
reshaped = random_tensor.view(3,2)

print(reshaped)
print(reshaped.size())
```
